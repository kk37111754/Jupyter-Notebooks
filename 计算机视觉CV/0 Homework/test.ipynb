```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# 创建一个宽512高512的黑色画布，RGB(0,0,0)即黑色
np.set_printoptions(threshold=5)
img = np.zeros((512,512,3),np.uint8)

# set colour
RoyalBlue = (65,105,225)

# create word
for i in range(2):
    cv2.line(img, (50+250*i,150), (200+250*i, 150), RoyalBlue, 5)
    cv2.rectangle(img,(75+250*i,200),(135+250*i,300),RoyalBlue,5)
    cv2.line(img, (175+250*i,150), (175+250*i, 375), RoyalBlue, 5)
    cv2.line(img, (130+250*i,350), (175+250*i, 375), RoyalBlue, 5)

# save graph
cv2.imwrite("res_word.jpg",img)

# using matplotlib show the graph
plt.imshow(cv2.imread('./res_word.jpg',cv2.IMREAD_UNCHANGED))
plt.title('picture-name')
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# 创建一个宽512高512的黑色画布，RGB(0,0,0)即黑色
np.set_printoptions(threshold=5)
img = np.zeros((512,775,3),np.uint8)

# set colour
RoyalBlue = (225,105,65)

# create word "小"
cv2.line(img, (175,120), (175, 400), RoyalBlue, 5)
cv2.line(img, (120,350), (175, 400), RoyalBlue, 5)
cv2.line(img, (100,225), (50, 300), RoyalBlue, 5)
cv2.line(img, (250,225), (300, 300), RoyalBlue, 5)

# create word "夏"
cv2.line(img, (450,125), (650, 125), RoyalBlue, 5)
cv2.line(img, (525,125), (475, 150), RoyalBlue, 5)
cv2.rectangle(img,(465,150),(635,300),RoyalBlue,5)
cv2.line(img, (465,200), (635, 200), RoyalBlue, 5)
cv2.line(img, (465,250), (635, 250), RoyalBlue, 5)
cv2.line(img, (530,300), (450, 340), RoyalBlue, 5)
cv2.line(img, (490,320), (630, 320), RoyalBlue, 5)
cv2.line(img, (630,320), (480, 420), RoyalBlue, 5)
cv2.line(img, (500,320), (630, 420), RoyalBlue, 5)


# save graph
cv2.imwrite("res_xia.jpg",img)

# using matplotlib show the graph
plt.imshow(cv2.imread('./res_xia.jpg',cv2.IMREAD_UNCHANGED))
plt.title('picture-name')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# open file
img = cv2.imread(r'./test158.png')

# get dimension of nose
nose = img[116:160, 309:350]

# set the nose on the graph
for x in range(2):
    for y in range(5):
        img[44*(x):44*(x+1),41*(y):41*(y+1)] = nose

# save graph
cv2.imwrite("res158.jpg",img)

# using matplotlib show the graph
plt.imshow(cv2.imread('./res158.jpg',cv2.IMREAD_UNCHANGED))
plt.title('picture-rgb')
plt.show()
```
